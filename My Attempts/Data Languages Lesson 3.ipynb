```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
from matplotlib import pyplot as plt
samplingFrequency = 16384 # Sampling frequency [Hz]  
T = 60*60*4 # Duration [s]

N = T*samplingFrequency # Total number of points 

# Time interval spaced with 1/sf
timeInterval = np.arange(0, T, 1./samplingFrequency)

noise = np.random.normal(size=timeInterval.shape)

plt.figure(1)
plt.plot(timeInterval,noise)
```

```python
psd = noise.psd(4)
plt.figure(1)
plt.plot(timeInterval,psd)
```
